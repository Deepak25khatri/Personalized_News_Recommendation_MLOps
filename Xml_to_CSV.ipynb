```python
import requests
import xml.etree.ElementTree as ET
import pandas as pd

# Fetch the XML content from the URL
url = "https://www.thehindu.com/life-and-style/fashion/feeder/default.rss"
response = requests.get(url)
xml_content = response.content

# Parse the XML content
root = ET.fromstring(xml_content)

# Extract relevant data
data = []
for item in root.findall(".//item"):
    title = item.find("title").text if item.find("title") is not None else None
    link = item.find("link").text if item.find("link") is not None else None
    description = item.find("description").text if item.find("description") is not None else None
    pub_date = item.find("pubDate").text if item.find("pubDate") is not None else None
    category = item.find("category").text if item.find("category") is not None else None

    data.append({
        "Title": title,
        "Link": link,
        "Description": description,
        "Publication Date": pub_date,
        "Category": category
    })

# Convert to DataFrame
df = pd.DataFrame(data)

# Save to CSV
csv_path = "/content/drive/MyDrive/Summer_group_project/the_hindu_news_5.csv"
df.to_csv(csv_path, index=False)

print(f"CSV file saved to {csv_path}")
```

```python
import pandas as pd

# List of CSV file paths
csv_files = ["/content/drive/MyDrive/Summer_group_project/the_hindu_news_1.csv",
             "/content/drive/MyDrive/Summer_group_project/the_hindu_news_2.csv",
             "/content/drive/MyDrive/Summer_group_project/the_hindu_news_3.csv",
             "/content/drive/MyDrive/Summer_group_project/the_hindu_news_4.csv",
             "/content/drive/MyDrive/Summer_group_project/the_hindu_news_5.csv"]

# Read each CSV file into a DataFrame and add it to a list
dataframes = [pd.read_csv(file) for file in csv_files]

# Concatenate all DataFrames in the list into a single DataFrame
merged_df = pd.concat(dataframes)

# Save the merged DataFrame to a new CSV file
merged_df.to_csv("/content/drive/MyDrive/Summer_group_project/merged_file_hindu_news_1.csv", index=False)

print("CSV files have been merged successfully!")
```
