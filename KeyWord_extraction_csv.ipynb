```python
import pandas as pd

def extract_keyphrases(text, n=10):
    extractor = pke.unsupervised.TfIdf()
    extractor.load_document(input=text, language='en')
    extractor.candidate_selection()
    extractor.candidate_weighting()
    keyphrases = extractor.get_n_best(n=n)
    return [phrase for phrase, score in keyphrases]

# Read the CSV file into a DataFrame
df = pd.read_csv("/content/drive/MyDrive/Summer_group_project/merged_file_hindu_news_1_updated_file.csv")

# Apply the keyphrase extraction to each description
df['Keyphrases'] = df['New_description'].apply(lambda x: ', '.join(extract_keyphrases(str(x))))

# Save the updated DataFrame back to a CSV file
df.to_csv("/content/drive/MyDrive/Summer_group_project/updated_file_with_keyphrases.csv", index=False)

print("Keyphrases have been extracted and added to the new column 'New_description' successfully!")
```
